### Multi D List

```python
l1 = [1,2,3,[4,5],[6,7,9]]
l1
```

```python
arr2 = np.array([[1,2,3],[4,5,7],[8,9,10]])
print(arr2)
```

```python
# one's Array
one = np.ones((2,3),dtype = int)
print(one)
```

```python
zero = np.zeros((2,3),dtype = int)
print(zero)
```

```python
identity = np.eye(3,dtype = int)
print(identity)
```
